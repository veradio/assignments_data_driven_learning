```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
import sklearn as sk
```

```python
#changed one smal part in the txt file, there was a return to much at 149. so i backspaced it. 
jokes = pd.read_csv("text jokes.txt", sep=":\t", engine='python', header=None).rename(columns={0: "joke_id", 1: "joke"})
jokes = jokes.set_index("joke_id")
jokes.head()
```

```python
ratings = pd.read_csv("data jokes.csv", header=None)
ratings.head()
```

```python
ratingszero = ratings.replace(regex=[99.0], value= 0.0) #changing the ratings of 9.9 to 0.0 so it won't influence the results that much, 
                                                        #beceause the other values are between -10 and 10.. 0.0 is exactly in the middle
                                                        # 99 would influence the clusters to much
ratingszero = ratingszero.dropna()
```

```python
from sklearn.cluster import KMeans
km = KMeans(n_clusters=4, random_state=1)
km = km.fit(ratingszero)
ratingszero["clusters"] = km.predict(ratingszero)
ratingszero["clusters"].value_counts()
```

KMeans measures the distance to other values in the data set, then it's searching for other values in the data set with the same distance from the centralized "cluster" point. With this clusters it's possible to see which groups are close to eachoter, and if there are differences inbetween the clusters. 

```python
#Using a dataframe with only the columns that contain no missing values, 
#perform a k-means clustering on users (experiment a bit to get a suitable number of clusters). 


mean_ratings = pd.pivot_table(ratingszero, index = "clusters")
mean_ratings
```

```python
mean_ratings = mean_ratings.reindex(ratings.count().sort_values(ascending=False).index, axis=1) 
turned = mean_ratings.transpose() #turning the dataframe with the clusters so it's possible to merge it lateron
mean_ratings
```

```python
#Investigate the clusters and try to give an interpretation with the joke texts: 
#what type of user are they? This may not always be easy/possible.
df = jokes.merge(turned, how='inner', left_index=True, right_index = True)
#merging the jokes to the cluster dataframe

cluster0 = df[df[0] != 0].sort_values(0, ascending=False).head(3)
print("jokes cluster 0:", cluster0["joke"])
print("")
cluster1 = df[df[1] != 0].sort_values(1, ascending=False).head(3)
print("jokes cluster 1:", cluster1["joke"])
print("")
cluster2 = df[df[2] != 0].sort_values(2, ascending=False).head(3)
print("jokes cluster 2:", cluster2["joke"])
print("")
cluster3 = df[df[3] != 0].sort_values(3, ascending=False).head(3)
print("jokes cluster 3:", cluster3["joke"])
```

The users of the following clusters do like:

Cluster 0: does like almost all the jokes
Cluster 1: doesn't like much jokes, but they like short jokes about stupid mistakes.
Cluster 2: close to cluster 1
Cluster 3: like shot oldskool jokes
