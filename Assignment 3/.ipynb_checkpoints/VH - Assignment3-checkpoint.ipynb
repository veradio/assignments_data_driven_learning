```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
import sklearn as sk
```

IBM HR Analytics Employee Attrition & Performance
Predict attrition of your valuable employees

Description as written on Kaggle: 
'Uncover the factors that lead to employee attrition and explore important questions such as ‘show me a breakdown of distance from home by job role and attrition’ or ‘compare average monthly income by education and attrition’. This is a fictional data set created by IBM data scientists.'

```python
df = pd.read_csv('Employees.csv')
df.info() #checking what kind of values there are in this dataframe and check if there's something missing
```

```python
correlation = df.corr() #check which values are strong correlated
plt.figure(figsize=(9,7))
sns.heatmap(correlation, linewidths=.5)
```

```python
print (df["StandardHours"].value_counts())
print (df["EmployeeCount"].value_counts())
```

Dropping the StandardHours and EmployeeCount, beceause those are the same for every row. Dropping the EmployeeNumber, beceause it's a given "index" number by the company which is not a value thats relating to Attrition.

Also dropping some variables that are strongly correlated. Beceause of the storng correlation, other data is already giving the same output".

```python
df = df.drop(["StandardHours", "EmployeeCount","EmployeeNumber", "JobLevel","TotalWorkingYears", "YearsAtCompany","YearsInCurrentRole","YearsSinceLastPromotion", "YearsWithCurrManager"], axis = 1)
df.head()
```

```python
Num_val = {'Yes':1, 'No':0} #changing the yes and no from attrition to 1 and 0, beceause when using supervised machine learning it's not possible to have string/object data
df['Attrition'] = df["Attrition"].apply(lambda x: Num_val[x])
df.head()
```

```python
df = pd.get_dummies(df) #collecting dummies, beceause for training the model all data should be in the same numerical format
df.head()
```

```python
from sklearn.model_selection import train_test_split 

y = df['Attrition'] #create the y-variable
X = df.drop("Attrition", axis = 1) #create the X matrix
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=1) #split the data, store it into different variables
```

```python
from sklearn.ensemble import RandomForestClassifier
model= RandomForestClassifier()
model = RandomForestClassifier()
model.fit(X_train,y_train)
```

I've used random forrest calssifier, beceause it creates many desiciontrees and tries to find the best combinations to predict the model on, this does randomforest by taking the mode of all the values in an descision tree. This is better than using a desicionstree, beceaus it looks also at the weak relations between variables, wat might be relevant in some cases with loads of data. 

```python
from sklearn.metrics import confusion_matrix
test_pred = model.predict(X_test)

y_p = model.predict(X_test) 
cm = confusion_matrix(y_test, y_p)
cm = pd.DataFrame(cm, index=['no attrition(actual)', 'attrition (actual)'], columns = ['no attrition (pred)', 'attrition (pred)']) #label and make df
cm
```

```python
recall = cm.iloc[1,1] / (cm.iloc[1,0] + cm.iloc[1,1]) 
precision = cm.iloc[1,1] / (cm.iloc[0,1] + cm.iloc[1,1])
print(f"The recall is {recall}.\nThe precision is {precision}")
```

this is not a really good prediction, in this prediction there's 80% missing of the empoyees that are going for attrition. The times where the model predicted it right is just only 71%.
