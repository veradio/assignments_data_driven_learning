```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import scipy.stats
from scipy.stats.stats import pearsonr
sns.set(rc={'figure.figsize':(11.7,8.27)})
```

```python
df1 = pd.read_csv('steps_data.csv', sep = ';')
df2 = pd.read_csv('survey_data.csv')
```

```python
#merge two dataframes on id, so it's one dataset
df = pd.merge(df2, df1, on = 'id')
df.head()
```

```python
#plot of a time series, e.g. mean steps per day over all participants
#df1 is the dataframe which contains the steps
meansteps = df1.drop("id",axis =1).mean()

dfmeansteps = pd.DataFrame(meansteps.dropna().astype(int))
dfmeansteps = dfmeansteps.iloc[1:328]
dfmeansteps.rename(columns ={0: 'steps'},inplace = True)
#dfmeansteps= dfmeansteps[(dfmeansteps["raw_character_text"] == "Lisa Simpson")

mediansteps = int(dfmeansteps["steps"].median())
print(f"The median steps: {mediansteps} per date ")

mean_meansteps = int(dfmeansteps["steps"].mean())
print(f"The variance in mean steps: {mean_meansteps} per date")
```

```python
dfmeansteps.index = pd.to_datetime(dfmeansteps.index) #converting index from str to dates, to create a time series plot

#checking on the maximum steps and the minimum steps
print("the least steps where taken on", dfmeansteps["steps"].idxmin(), "with a mean of" , dfmeansteps["steps"].min(), "Steps." )
print("the most steps where taken on", dfmeansteps["steps"].idxmax(), "with a mean of" , dfmeansteps["steps"].max(), "Steps." )
```

```python
#A plot of a time series of your choice, e.g. mean steps per day over all participants
plt.figure(figsize=(9,5))

sns.lineplot(data=dfmeansteps)
sns.set_style("darkgrid")
plt.show()
```

There is one Outliner, that's on 2013-12-25 which is Crhistmas day. On that date was christmas day. 

```python
#2 graphs of the distribution (e.g., histogram, box plot, density plot, violin plot)
sns.distplot(dfmeansteps.dropna(), norm_hist= True )
plt.axvline(int(mediansteps), color='green', label='median')
plt.axvline(int(mean_meansteps), color='red', label='mean')
plt.xlabel('Mean Steps Per Day')
plt.ylabel('Participants')
plt.legend()
plt.show()
```

the normality 

```python
#2 graphs of the distribution (e.g., histogram, box plot, density plot, violin plot)

sns.violinplot(dfmeansteps)
plt.axvline(int(mediansteps), color='green', label='median')
plt.axvline(int(mean_meansteps), color='red', label='mean')
plt.title('average daily steps per participants')
plt.xlabel('Average Steps per Day')
plt.ylabel('Participants')
plt.show()
```

```python
#create a different id dataframe, 
#so it's possible to get the mean steps per paricitpant without adding the id
col_id = df1["id"]
df_id = pd.DataFrame(col_id)
```

```python
#create a different days dataframe, 
#so it's possible to get the mean steps per paricitpant
days = df1.drop("id", axis =1) 
dfsteps = pd.DataFrame(days.mean(axis = 1)).dropna().astype(int)
```

```python
#merging the id dataframe and dfsteps, so it's possible to get the mean steps per id
df_steps_participant =  df_id.join(dfsteps)
df_steps_participant.head()
```

```python
#merging the steps per paricitpant to the survey dataframe
#so it's possible to get the mean steps and the additional survey data for each participant
df3 = pd.merge(df2, df_steps_participant, on = 'id').dropna()
df3.rename(columns ={0: 'mean_steps'},inplace = True) 
```

```python
#removing all steps above 20000 beceause that's seems impossible for a mean steps per day
df3 = df3[(df3["mean_steps"] <=20000)]

#removing all waist sizes above 127cm beceause the biggest clothing size for men XXXL goes upto 127cm
#removing all waist sizes below 55, beceause that's impossible womens size xxs is 60cm waist
df3 = df3[(df3["waist"] <= 127) & (df3["waist"] >= 55)]
```

```python
#Correlation matrix including all 3 variables
sns.pairplot(df3[["weight","waist","mean_steps"]])
```

There is a strong positive liniear relation between waist and weight, what makes sense beceause the more your weight the bigger the wast. 

Incombination with mean_steps the waist and weight show both weak negative relationships.

All of the plots in the matrix are a little bit right skewed. 

```python
#Scatter plots of your variable with the 2 other variables (2 scatter plots in total)
plt.figure(figsize=(9,5))
sns.scatterplot(x="mean_steps",y="weight", hue = "waist", data=df3)
```

```python
#Scatter plots of your variable with the 2 other variables (2 scatter plots in total)
#Pearsonâ€™s correlation for the 2 relations
plt.figure(figsize=(9,5))
sns.scatterplot(x="mean_steps",y="weight", data=df3)
print(pearsonr(df3['mean_steps'],df3['weight']))
plt.show
```

mean_steps and weight are showing a weak negative correlation R = -0.152 but the significance is high P = 0.0000572 

```python
#Pearsonâ€™s correlation for the 2 relations
plt.figure(figsize=(9,5))
sns.scatterplot(x="mean_steps", y = "waist", data=df3)
print(pearsonr(df3['mean_steps'],df3['waist']))
plt.show
```

mean_steps and waist are showing a weak negative correlation R = -0.107 but the significance is high P = 0.004

```python
#The relation of mean steps per participant 
#with a qualitative variable with two levels (e.g. gender, weekday vs weekend)
plt.figure(figsize=(12,9))
sns.catplot(x='gender',y='mean_steps', kind="swarm",data=df3)
plt.title('Mean steps per day based on gender')
plt.ylabel('Mean Steps Per Day')
plt.xlabel('Gender')
plt.show()
```

```python
#The result of a independed t-test
from scipy.stats import ttest_ind 


male = df3[df3['gender']== "male"]
female = df3[df3['gender']== "female"]


ttest_ind(male['mean_steps'], female['mean_steps'])
```

The P value = 0.81 which is much higher than 0.05 (significance of 95%) which makes it possible to conclude, that the differences between the meansteps of males and the meansteps of females are not showing a big difference.
